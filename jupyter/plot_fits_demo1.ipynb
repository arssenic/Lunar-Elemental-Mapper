```python
from astropy.io import fits

# Open the FITS file
with fits.open(fits_file_path) as hdul:
    # Access the HDU that contains the spectral data
    spectrum_hdu = hdul[1]  # Adjust the index if needed
    data = spectrum_hdu.data

    # Print the names of the columns (for confirmation)
    print("Available columns in the data:")
    print(data.columns.names)

    # Extract the 'CHANNEL' and 'COUNTS' columns
    channel_data = data['CHANNEL']
    counts_data = data['COUNTS']  # Use 'COUNTS' as the value column

    # Create a dictionary to hold channel-count pairs
    channel_value_map = {}

    # Populate the dictionary with channel and corresponding count
    for channel, count in zip(channel_data, counts_data):
        if channel not in channel_value_map:  # Ensure each channel only gets one value
            channel_value_map[channel] = count

    # Print the values corresponding to each channel from 0 to 2047
    print("Channel Values (0-2047):")
    for channel in range(2048):  # Loop from 0 to 2047
        if channel in channel_value_map:
            print(f"Channel {channel}: {channel_value_map[channel]}")
        else:
            print(f"Channel {channel}: No Data")  # If no data exists for that channel
```

```python
import os
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt

# Define the directory containing the FITS files
data_directory = '/mnt/c/Users/r9307/interIIT/ch2_class_pds_release_38_20240927/cla/calibration/'


# Create a dictionary to hold channel and non-zero count values for all files
all_channel_values = {}

# Initialize the plot
plt.figure(figsize=(15, 10))

# Counter for subplot indexing
subplot_index = 1

# Iterate through all files in the specified directory
for filename in os.listdir(data_directory):
    if filename.endswith('.fits'):
        fits_file_path = os.path.join(data_directory, filename)
        print(f"Processing file: {fits_file_path}")
        
        # Open the FITS file
        with fits.open(fits_file_path) as hdul:
            # Check the number of HDUs in the file
            for i in range(len(hdul)):
                # Access the HDU that contains the spectral data
                if hasattr(hdul[i], 'data') and hdul[i].data is not None:
                    spectrum_data = hdul[i].data

                    # Extract the 'CHANNEL' and 'COUNTS' columns
                    channel_data = spectrum_data['CHANNEL']
                    counts_data = spectrum_data['COUNTS']

                    # Store non-zero counts in a dictionary
                    for channel, count in zip(channel_data, counts_data):
                        if count > 0:  # Store only non-zero counts
                            if channel not in all_channel_values:
                                all_channel_values[channel] = []
                            all_channel_values[channel].append(count)

                    # Plot the spectrum in a subplot
                    plt.subplot(2, 1, subplot_index)
                    plt.plot(channel_data, counts_data, label=filename, alpha=0.5)
                    plt.xlabel('Channel Number')
                    plt.ylabel('Counts')
                    plt.title('X-ray Spectrum from Chandrayaan-2')
                    plt.legend(loc='upper right')
                    plt.grid()
                    subplot_index += 1
                    
                    # Limit subplot index to avoid exceeding number of subplots
                    if subplot_index > 2:
                        plt.tight_layout()
                        plt.show()
                        plt.figure(figsize=(15, 10))  # Create a new figure for more plots
                        subplot_index = 1

# Finalize the plots
if subplot_index > 1:
    plt.tight_layout()
    plt.show()

# Display the non-zero channel values
print("Non-zero channel values for each file:")
for channel, counts in all_channel_values.items():
    print(f"Channel {channel}: {counts}")
```
