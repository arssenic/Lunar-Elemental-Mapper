```python
import os
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from scipy.ndimage import gaussian_filter1d
from scipy.signal import find_peaks
from scipy.optimize import curve_fit

# Gaussian function for fitting
def gaussian(x, a, mu, sigma):
    return a * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Known K energies (in keV) for element identification
ka_energies = {
    "Mg": [1.25, 0.1],
    "Al": [1.49, 0.1],
    "Si": [1.74, 0.135],
    "Ca": [3.69, 0.1],
    "Fe": [6.40, 0.1]
}

def moving_average(data, window_size):
    """Compute the moving average using a specified window size."""
    return np.convolve(data, np.ones(window_size) / window_size, mode='same')

def identify_element(mu, tolerance=0.1):
    """
    Identifies the element based on the detected mu value and known K energies.

    Parameters:
    - mu (float): The detected energy of the peak in keV.
    - tolerance (float): The range within which to consider a match.

    Returns:
    - str: The identified element or 'Unknown' if no match is found.
    """
    for element, energy in ka_energies.items():
        if abs(mu - energy[0]) <= energy[1]:
            return element
    return "Unknown"

def process_batch(data_directory, batch_files, background_counts=None, window_size=10):
    """
    Processes a batch of .fits files, calculates smoothed counts, plots the spectrum,
    identifies peaks, and maps elements.

    Parameters:
    - data_directory (str): Path to the directory containing .fits files.
    - batch_files (list): List of .fits files in the current batch.
    - background_counts (np.array, optional): Background counts for subtraction.
    - window_size (int, optional): Window size for the moving average.
    """
    # Initialize counts array for summing up counts in this batch
    summed_counts = np.zeros(2048)
    # Process each file in the batch
    for filename in batch_files:
        file_path = os.path.join(data_directory, filename)
        with fits.open(file_path) as hdul:
            spectrum_data = hdul[1].data
            channels = spectrum_data['CHANNEL']
            counts = spectrum_data['COUNTS']
            summed_counts += counts

    # Calculate energy for each channel
    energy = np.array([13.58 * c / 1000 for c in channels])
    summed_counts_s = np.array([c / 12 for c in summed_counts])

    # Filter channels between 37 and 800
    valid_indices = (channels >= 50) & (channels <= 300)
    channels = channels[valid_indices]
    energy = energy[valid_indices]
    summed_counts_s = summed_counts_s[valid_indices]

    # Perform background subtraction if background counts are available
    if background_counts is not None:
        background_counts = background_counts[valid_indices]
        summed_counts_s -= background_counts

    # Compute moving average on the filtered summed_counts
    smoothed_counts = moving_average(summed_counts_s, window_size)

    # Plot raw counts and smoothed (moving average) counts vs. Energy
    plt.figure(figsize=(10, 6))
    plt.plot(energy, smoothed_counts, label=f'{window_size}-Channel Moving Average', color='red', linewidth=2)

    # Configure plot labels and title
    plt.xlabel('Energy (keV)')
    plt.ylabel('Counts/s')
    # plt.yscale('log')
    plt.title('X-ray Spectrum with Moving Average Smoothing')
    plt.legend(loc='upper right', fontsize='small')
    plt.grid()
    plt.tight_layout()
    plt.show()

    # Identify and fit peaks for the current batch
    identify_and_fit_peaks(energy, smoothed_counts)

def identify_and_fit_peaks(energy, smoothed_counts, window_size=5):
    """
    Identifies peaks, fits a Gaussian curve to each peak, and matches the peak energy to an element.

    Parameters:
    - energy (np.array): Energy values in keV.
    - smoothed_counts (np.array): Smoothed counts data.
    - window_size (int): Number of data points around the peak to consider for fitting.
    """
    # Smooth the counts data
    smoothed_counts = gaussian_filter1d(smoothed_counts, sigma=2)

    # Calculate threshold based on mean and standard deviation
    mean_count = np.mean(smoothed_counts)
    std_count = np.std(smoothed_counts)
    threshold = mean_count + std_count

    # Identify peaks
    peaks, _ = find_peaks(smoothed_counts, height=threshold, distance=10)
    peaks_info = []

    # Plot for visualizing peaks and fits
    plt.figure(figsize=(10, 6))
    plt.plot(energy, smoothed_counts, label='Smoothed Data', color='blue')

    for peak in peaks:
        # Select data around the peak for fitting
        start = max(0, peak - window_size)
        end = min(len(energy), peak + window_size)
        energy_window = energy[start:end]
        counts_window = smoothed_counts[start:end]

        # Initial guess for Gaussian parameters: amplitude, mean, and standard deviation
        initial_guess = [np.max(counts_window), energy[peak], 0.1]

        try:
            # Perform Gaussian fit
            popt, _ = curve_fit(gaussian, energy_window, counts_window, p0=initial_guess)
            a, mu, sigma = popt

            # Identify the element based on the mu value
            element = identify_element(mu)

            # Store peak info (mu, sigma, and identified element)
            peaks_info.append((mu, sigma, element))

            # Plot the fitted Gaussian
            fitted_curve = gaussian(energy_window, *popt)
            plt.plot(energy_window, fitted_curve, label=f'Peak at {mu:.2f} keV ({element})', linestyle='--')
            plt.axvline(mu, color='red', linestyle=':', alpha=0.7)
        except RuntimeError:
            print(f"Gaussian fit failed for peak at energy index {peak}")

    # Plot configuration
    plt.xlabel('Energy (keV)')
    plt.ylabel('Counts/s')
    # plt.yscale('log')
    plt.title('X-ray Spectrum with Gaussian Fits and Element Identification')
    plt.legend()
    plt.grid()
    plt.show()

    # Print peak information
    print("Peak Information:")
    for i, (mu, sigma, element) in enumerate(peaks_info):
        print(f"  Peak {i+1}: mu = {mu:.2f} keV, sigma = {sigma:.2f} keV, Element = {element}")

# Example usage:
data_directory = '/mnt/c/Users/r9307/interIIT/feb1fits_files/feb1fits_files'
background_file = '/mnt/c/Users/r9307/interIIT/high_solar_angle_90output_unranged_addedkernels_avg_back.fits'
with fits.open(background_file) as hdul:
    spectrum_data = hdul[1].data
    bg_counts = spectrum_data['COUNTS']
    print(bg_counts.shape)
batch_size = 12
window_size = 10
tolerance = 0.1

# Get all .fits files in the directory
fits_files = [f for f in os.listdir(data_directory) if f.endswith('.fits')]
num_batches = len(fits_files) // batch_size

for batch_idx in range(10,25):
    batch_files = fits_files[batch_idx * batch_size: (batch_idx + 1) * batch_size]
    print(f"\nProcessing Batch {batch_idx + 1}...")
    process_batch(data_directory, batch_files, bg_counts, window_size=window_size)
```
